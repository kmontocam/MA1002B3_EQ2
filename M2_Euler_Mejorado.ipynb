In Problems 1, 2, 3, 4, 5, 6, Z. 8, 9, and 10 use the improved Euler's method to obtain a four-decimal approximation of the indicated value. First use $h = 0.1$ and then use $h= 0.05$.

```python
import numpy as np
import pandas as pd
import sympy as sp
from IPython.display import display, Math, Latex

x = sp.symbols('x')
g = sp.Function('g')
```

## Función que resuelve ecuaciones diferenciales a través del método de Euler mejorado

```python
def edo_solver_improved_euler_method(edo, a: int, b: int, y: int, n: int):
    
    interval, delta = np.linspace(a, b, n + 1, endpoint = True, retstep = True)
    
    for x_n_1, x_n in zip(interval, interval[1:]):
        y = y + delta * (edo(x_n_1, y) + edo(x_n, y + delta * edo(x_n_1, y)))/2
    return y
```

## Función que regresa tabla comparando la solución a través del método de Euler mejorado con una solución analítica

```python
def comparison_table_improved_euler_method_and_analytic_method(edo, a: int, b: int, y: int, n: int, x_def: int, diffeq):
    
    interval, delta = np.linspace(a, b, n + 1, endpoint = True, retstep = True)
    
    sol = sp.dsolve(diffeq, g(x), ics = {g(x_def): y})
    
    tb = pd.DataFrame({'xn': [a], 'yn': [y], 'y': [y], '|y - yn|': [0], '|(y - yn)/y| * 100%': [0]})
    tb.index.name = 'n'
    
    for i, [x_n_1, x_n] in enumerate(zip(interval, interval[1:])):
        
        y = y + delta * (edo(x_n_1, y) + edo(x_n, y + delta * edo(x_n_1, y)))/2
        
        y_ex = (sol.evalf(subs = {x: x_n}).rhs)
        
        abs_error = abs(y_ex - y)
        
        rel_error = abs((y_ex - y)/y_ex) * 100
        
        new_row = {'xn': x_n, 'yn': y, 'y': y_ex, '|y - yn|': abs_error, '|(y - yn)/y| * 100%': rel_error}
        tb = tb.append(new_row, ignore_index = True)
    
    return tb
```

### Función que convierte h a número de divisiones

```python
def h_to_n(a, b, h):
    n = (b - a)/h
    return int(n)
```

## 1. $y' = 2x -3y +1, y(1) = 5; y\left(\frac{3}{2}\right)$

```python
edo_1 = lambda x, y: 2*x - 3*y + 1
diffeq_1 = sp.Eq(sp.Derivative(g(x), x), (2 * x) - (3 * g(x)) + 1)
```

### Solución con $h = 0.1$

```python
n = h_to_n(a = 1, b = 1.5, h = 0.1)
comparison_table_improved_euler_method_and_analytic_method(edo_1, 1, 1.5, 5, n, 1, diffeq_1)
```

### Solución con $h = 0.05$

```python
n = h_to_n(a = 1, b = 1.5, h = 0.05)
comparison_table_improved_euler_method_and_analytic_method(edo_1, 1, 1.5, 5, n, 1, diffeq_1)
```
